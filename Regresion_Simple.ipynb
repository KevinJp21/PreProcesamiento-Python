# Práctica 3: Regresión lineal simple

1. cargar librerias
2. Generar datos
3. Graficar
4. Modelo de regresion
5. Predecir

## 1. Cargar librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## 2. Generar datos

```python
x = [random.uniform(0,10) for _ in range (50)]
y = [2 * xi + 1 + random.uniform(-1,1) for xi in x]
```

## 3. Graficar datos

```python
plt.scatter(x,y, color='blue', label='Datos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Datos aleatorios')
plt.legend()
plt.show()
```

## 4. Modelo

```python
x_media = np.mean(x)
y_media = np.mean(y)

numerador = np.sum((x-x_media) * (y-y_media))
denominador = np.sum((x-x_media) **2)

a=numerador/denominador
print('La pendiente a es: ',a)

b=y_media - a*x_media
print('El intercepto b es: ',b)
```

## 5. Prediccion

```python
y_pred = [a*xi + b for xi in x]
```

```python
plt.scatter(x,y, color='blue', label='Datos')
plt.plot(x,y_pred, color = 'red', label='Regresion')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Datos aleatorios')
plt.legend()
plt.show()
```
