## 1. Cargar librerias

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## 2. Leer datos

```python
# Forma 1
# Generar datos aleatorios
data = np.random.uniform(0, 50, (100,2))
```

```python
# Forma 2
#Definiendo número de muestras, y Rangos Min y Max de los datos

num_samples = 100
data_range_min = 0
dat_range_max = 50

data = np.random.uniform(low=data_range_min, high=dat_range_max, size=(num_samples, 2))
```

## 3. Kmeans

```python
# Aplicar el algoritmo KMeans
k= 5

#Definir algoritmo
kmeans = KMeans(n_clusters = k)

#Entrenar modelo
kmeans.fit(data)

# Usar modelo
labels = kmeans.predict(data)

#Mostrar centroides
centers = kmeans.cluster_centers_
```

## 4. Gráficar clusters

```python
# Gráficar los clústers
plt.figure(figsize = (8,6))
plt.scatter(data[:, 0], data [:, 1], c=labels, cmap = 'viridis')
plt.scatter(centers[:, 0], centers[:, 1], c= 'red', marker = '*')
plt. title(f'Clustering con KMeans (k = {k})')
plt.xlabel('X')
plt.ylabel('Y')
plt.show
```

```python
from ipywidgets import interact, IntSlider

def generate_random_data(num_samples, data_range):
    return np.random.uniform(low=data_range[0], high=data_range[1], size=(num_samples, 2))

def kmeans_clustering(k, num_samples, data_range_min, data_range_max):
    data_range = (data_range_min, data_range_max)
    
    # Generar datos aleatorios
    data = generate_random_data(num_samples, data_range)
    
    # Aplicar el algoritmo K-means
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    labels = kmeans.predict(data)
    centers = kmeans.cluster_centers_
    
    # Graficar los clústeres
    plt.figure(figsize=(8, 6))
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='*')
    plt.title(f'K-means Clustering (k={k})')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

# Función de interacción para ajustar los parámetros
interact(kmeans_clustering, 
         k =IntSlider(min=1, max=10, step=1, value=3, description='K:'),
         num_samples=IntSlider(min=50, max=10000, step=50, value=100, description='N:'),
         data_range_min=IntSlider(min=0, max=10, step=1, value=0, description='Data Min:'),
         data_range_max=IntSlider(min=0, max=100, step=10, value=50, description='Data Max:'))
```
